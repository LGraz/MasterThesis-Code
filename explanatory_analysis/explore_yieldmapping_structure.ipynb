```python
import pandas as pd
import os

os.chdir(r"/home/lukas/Documents/ETH/MASTER_THESIS/code")
data_dir = r"data/yieldmapping_data"

## This takes a data-directory with subdirectories
## and only csv-files.
## Lists all collumn-names of the files like:
##      name set A
##        file1
##        file2
##
##      name set B
##        file3
##        file4
def list_collumn_names(data_dir):
    legend_items = [
        "FID	cum_rain	day_rain	avg_temp	max_temp	min_temp	gdd	das	date",
        "coord_id	dry_yield	x_coord	y_coord	FID	epsg	harvest_year",
        "coord_id	B02	B03	B04	B05	B06	B07	B08	B8A	B11	B12	cum_rain	avg_temp	day_rain	max_temp	min_temp	gdd	das	date	FID	scene_id	product_uri	x_coord	y_coord	scl_class	epsg	harvest_year"
    ]
    #print("data dir: " + data_dir)
    listdir_out = os.listdir(data_dir)
    dirs = []
    files = []
    for item in listdir_out:
        item_path = os.path.join(data_dir, item)
        if os.path.isdir(item_path):
            dirs.append(item_path)
        elif os.path.isfile(item_path):
            files.append(item_path)
        else:
            print("some error")

    ind = []
    for i, path in enumerate(files):
        df = pd.read_csv(path)
        df_names = "\t".join(df.columns)
        # print(df_names)
        try:
            ind.append(legend_items.index(df_names))
        except:
            print("ERROR: " + df_names)
            print("Path= " + path)

    for i, name in enumerate(legend_items):
        print("\n" + name)
        for j, file in enumerate(files):
            if ind[j] == i:
                print(file)

    for dir in dirs:
        print("\n" + "--- "+dir+" ---")
        list_collumn_names(dir)
        print("------- end dir --------")


list_collumn_names(data_dir)
```
